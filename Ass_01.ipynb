```python
df = df.query("pickup_latitude>=-90 and pickup_latitude <=90")
df = df.query("dropoff_latitude>=-90 and dropoff_latitude <=90")
df = df.query("pickup_longitude>=-180 and pickup_longitude <=180")
df = df.query("dropoff_longitude>=-180 and dropoff_longitude <=180")
```

```python
df
```

```python
import math

def haversine(lat1, lon1, lat2, lon2):
    lat1, lon1, lat2, lon2  = map(math.radians, [lat1, lon1, lat2, lon2])
    dlon = lon2-lon1
    dlat = lat2-lat1
    a = math.sin(dlat/2)**2 + math.cos(lat1) * math.cos(lat2) * math.sin(dlon/2)**2
    c=2*math.asin(math.sqrt(a))
    dkm = 6371*c
    return dkm
```

```python
df["distance_km"] = df.apply(lambda row: haversine(row["pickup_latitude"], row["pickup_longitude"], row["dropoff_latitude"], row["dropoff_longitude"]), axis = 1)
```
